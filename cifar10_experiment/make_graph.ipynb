```python
import matplotlib.pyplot as plt
%matplotlib inline

import pickle

# plt.style.use("fivethirtyeight")
plt.style.use("bmh")
```

```python
x = range(1,51)

with open('history_lenet.dump', 'rb') as f:
    lenet = pickle.load(f)

with open('history_alexnet.dump', 'rb') as f:
    alexnet = pickle.load(f)

with open('history_vggnet.dump', 'rb') as f:
    vggnet = pickle.load(f)

with open('history_resnet.dump', 'rb') as f:
    resnet = pickle.load(f)
```

```python
plt.plot(x, lenet['val_acc'], label='LeNet')
plt.plot(x, alexnet['val_acc'], label='AlexNet')
plt.plot(x, vggnet['val_acc'], label='VGGNet')
plt.plot(x, resnet['val_acc'], label='ResNet')



plt.xlim([0,50])
plt.ylim([0,1.0])
plt.xlabel('epoch')
plt.ylabel('accracy')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0)
plt.savefig('graphs/accracies', format='pdf')
plt.show()
```

```python
plt.plot(x, lenet['val_loss'], label='LeNet')
plt.plot(x, alexnet['val_loss'], label='AlexNet')
plt.plot(x, vggnet['val_loss'], label='VGGNet')
plt.plot(x, resnet['val_loss'], label='ResNet')

plt.xlim([0,50])
plt.ylim([0,2.0])
plt.xlabel('epoch')
plt.ylabel('loss')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0)
plt.savefig('graphs/losses.pdf', format='pdf')
plt.show()
```

```python
plt.plot(x, lenet['acc'], label='train accuracy')
plt.plot(x, lenet['val_acc'], label='validation accuracy')

plt.xlim([0,50])
plt.ylim([0,1.0])
plt.xlabel('epoch')
plt.ylabel('accuracy')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0)
plt.savefig('graphs/lenet_acc.pdf', format='pdf')
plt.show()
```
